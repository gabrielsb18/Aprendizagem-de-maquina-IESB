IMPORTANDO ARQUIVO

```python
import pandas as pd #IMPORTANDO BIBLIOTECA PANDAS

df = pd.read_csv('dados4.csv', index_col='pclass') #lENDO ARQUIVO IMPORTADO
print(df.to_string()) #EXIBINDO TODAS AS MINHAS COLUNAS DO DATAFRAME
```

CODIGO PARA A QUESTAO 01

```python
df.dropna(subset = ['age'], inplace = True)
# O metodo ".dropna", dropa todas as linhas vazias, garantindo assim dados mais consistentes
df['age'] = df['age'].astype(int) # O metodo ".astype" est convertendo minha coluna idade de float para INT, para melhor visualizao

moda_da_idade = df['age'].mode()
#Com o metodo ".mode()" determinamos o valor que mais se repete, sendo a moda.
#Em "['idade']" estamos passando a coluna que queremos e extrair a informao.
print(moda_da_idade)

df.to_csv('Resposta01.txt')
#O metodo "to_csv" ser o responsvel por gerar nosso arquivo
```

CDIGO PARA RESPONDER A QUESTO 2

```python
sex=df['sex'].value_counts().to_string()
gender=sex.split()
print('2) Nmero de homens=', gender[1],', Nmero de mulheres =', gender[3])
print('Somatrio entre homens e mulheres=',(len(df.sex)))
```

CDIGO PARA RESPONDER A QUESTO 3

```python
import matplotlib.pyplot as plt #importando biblioteca Matplot para grficos

survivors=df['survived'].value_counts().to_string()
survivorsSplit=survivors.split()
print('3)Nmero de mortos =',survivorsSplit[1], 'Nmero de sobreviventes =', survivorsSplit[3])
print()

#CDIGO DO GRFICO DE PIZZA
labels = ['No-sobreviventes', 'Sobreviventes']
sizes = [809,500] #dados a serem computados pelo grfico

plt.figure(figsize=(5, 5)) #definio do tamanho do grfico a ser exibido
plt.pie(sizes, labels=labels, autopct='%1.1f%%', startangle=140) #definio da aparncia do grfico de pizza
plt.axis('equal')
plt.title('Nmero de sobreviventes e mortos')
plt.savefig('Resposta03.png') #gerar um arquivo que mostre o grfico
plt.show() #comando para exibir o grfico
print()
```
